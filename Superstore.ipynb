```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv(r"C:\Users\Smiley\Downloads\Sample - Superstore.csv")

```

```python
data.head(5)
```

```python
label_encoder = LabelEncoder()
```

```python
label_encoded_columns = ['Ship Mode', 'Segment']
label_mappings = {}
```

```python
for col in label_encoded_columns:
    data[col] = label_encoder.fit_transform(data[col])
    label_mappings[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
```

```python
print("Encoded Data:\n", data[label_encoded_columns].head())
print("\nLabel Mappings:\n", label_mappings)
```

```python
data['Ship Mode'] = label_encoder.fit_transform(data['Ship Mode'])
data['Segment'] = label_encoder.fit_transform(data['Segment'])
```

```python
data_encoded = pd.get_dummies(data, columns=['Region', 'Category', 'Sub-Category'])
```

```python
print(data_encoded.head())
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
